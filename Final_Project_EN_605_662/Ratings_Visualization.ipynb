```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('models_score.csv')
#print(df)
```

```python
df.head()
```

```python
def average_score_real(scores):
    size = len(scores)
    sum = 0

    for score in scores:
        sum += score

    average = sum / size
    return average
```

```python
# loop through the rows using iterrows()

models = []
average_ratings = []

for index, row in df.iterrows():

    print(row['Model'])
    models.append(row['Model'])

    scores = [row['Response1'], row['Response2'],
          row['Response3'], row['Response4'], row['Response5'],
          row['Response6'], row['Response7'], row['Response8'],
          row['Response9'], row['Response10'], row['Response11'],
          row['Response12'], row['Response13'], row['Response14'],
          row['Response15'], row['Response16'], row['Response17'],
          row['Response18'], row['Response19']]


    average = average_score_real(scores)
    average_ratings.append([average, scores])

    print(average)

    df.at[index, 'Average'] = average

#print(models)
#print(average_ratings)
```

```python
models[0] = "125m"
models[1] = "2.7b"
models[2] = "117m"
models[3] = "345m"
models[4]= "762m"
models[5]= "3b-lm"
models[6]= "175b"
models[7]= "13b"
models[8]="7b"
```

```python
dict_models = {}

for i in range(len(models)):
    dict_models[models[i]] = [average_ratings[i][0], average_ratings[i][1]]
```

```python
sort_models = ['117m', '125m', '345m', '762m' , '2.7b', '3b-lm', '7b', '13b', '175b']
sort_average_ratings = []
sort_ratings = []

for model in sort_models:
    sort_average_ratings.append(dict_models[model][0])
    sort_ratings.append(dict_models[model][1])

print(sort_average_ratings)
#print(sort_ratings)
```

```python
# Initialise the subplot function using number of rows and columns
figure, axis = plt.subplots(3, 3)

#['117m', '125m', '345m', '762m' , '2.7b', '3b-lm', '7b', '13b', '175b']
list_plot = [(0,0), (0,1), (0,2),(1,0), (1,1), (1,2), (2,0), (2,1), (2,2) ]

#plt.title('Box Plots of LLMs Responses Ratings', fontsize=14)

for i in range(len(sort_models)):
    axis[list_plot[i][0], list_plot[i][1]].boxplot(sort_ratings[i])
    axis[list_plot[i][0], list_plot[i][1]].set_title(sort_models[i]+" Model")


plt.show()
```

```python
# Visualizing The Score Average using Bar Charts
colors = ['green', 'blue', 'purple', 'brown', 'teal']
plt.bar(sort_models, sort_average_ratings, color=colors )
plt.title('Models Vs Average Ratings', fontsize=14)
plt.xlabel('Models', fontsize=14)
plt.ylabel('Average Ratings', fontsize=14)



plt.show()
```
