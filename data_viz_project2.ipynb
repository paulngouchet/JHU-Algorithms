```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
path = './adult/adult.data'
df = pd.read_csv(path, index_col=False)
```

```python

df.rename(columns={' <=50K': 'Income Group', '39': 'Age', ' Male':'Sex'}, inplace=True)
```

```python
df.rename(columns={' State-gov': 'Workclass', ' Adm-clerical': 'Occupation'}, inplace=True)
```

```python
df.rename(columns={' Bachelors': 'Education', ' 0': 'Capital Loss',  ' 2174': 'Capital Gain'}, inplace=True)
```

```python
df.rename(columns={' Never-married': 'Marital Status', ' White': 'Race'}, inplace=True)
```

```python
print(df.columns)
```

```python
df.pivot(columns='Income Group', values='Age').plot.hist()
plt.title("Histogram Age by Income Group")
plt.xlabel("Age")
plt.show()
```

```python
# Pivot the DataFrame to get the count of each income group within each workclass
pivot_df = df.pivot_table(index='Workclass', columns='Income Group', aggfunc='size', fill_value=0)

# Plot the bar chart
pivot_df.plot(kind='bar', stacked=True)
plt.title("Bar Plot: Workclass by Income Group")
plt.xlabel("Workclass")
plt.ylabel("Count")
plt.legend(title='Income Group')
plt.show()
```

```python
# Plot histogram
plt.hist(df['Capital Gain'], edgecolor='black')

# Set plot title and labels
plt.title("Histogram of Capital Gains")
plt.xlabel("Capital Gain")
plt.ylabel("Frequency")

# Display the histogram
plt.show()
```

```python
# Set the style of the plot
sns.set(style="whitegrid")

# Create a box plot using Seaborn
sns.boxplot(x='Capital Gain', data=df)

# Set plot title and labels
plt.title("Box Plot of Capital Gains")
plt.xlabel("Capital Gain")
plt.ylabel("Values")

# Display the box plot
plt.show()
```
